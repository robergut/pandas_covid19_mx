```python
# Condiciones
(ds["FECHA_DEF"] != "9999-99-99").value_counts()
```

```python
positivo_covid = (ds["CLASIFICACION_FINAL"]==1) | (ds["CLASIFICACION_FINAL"]==2) | (ds["CLASIFICACION_FINAL"]==3)
```

```python
confirmados = ds[positivo_covid]
```

```python
confirmados["FECHA_DEF"].unique()
```

```python
confirmados["FECHA_DEF"] != "9999-99-99"
```

```python
defunciones_covid = confirmados["FECHA_DEF"] != "9999-99-99"
```

```python
defunciones_covid.value_counts()
```

```python
fecha_sintomas = pd.to_datetime(confirmados["FECHA_SINTOMAS"])
```

```python
fecha_base = pd.to_datetime(today)
```

```python
diferencia_dias = fecha_base - fecha_sintomas
```

```python
diferencia_dias
```

```python
diferencia_dias.value_counts()
```

```python
intervalo_activo = pd.Timedelta("14 days")
intervalo_activo
```

```python
casos_activos = diferencia_dias < intervalo_activo
```

```python
casos_activos.value_counts()
```

```python
confirmados = confirmados[casos_activos]
```

```python
confirmados
```

```python
confirmados["ENTIDAD_RES"]
```

```python
catalogos = pd.read_excel("descriptores/201128 Catalogos.xlsx", sheet_name=None)
```

```python
catalogos.keys()
```

```python
estados = catalogos["Catálogo de ENTIDADES"]
estados
```

```python
activos_estado = pd.merge(estados, confirmados, how="right", left_on="CLAVE_ENTIDAD", right_on="ENTIDAD_RES")
activos_estado
```

```python
activos_estado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos confirmados activos")
```

```python
activos_estado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos activos").reset_index().rename(columns={"index":"Estado"})
```

```python
activos = activos_estado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos confirmados").reset_index().rename(columns={"index":"Estado"})
```

```python
p = bpl.figure(x_range=activos["Estado"], plot_width=800, plot_height=600,title="Casos activos por entidad federativa de residencia",y_range=(0,5000))
```

```python
colores = ["#800000", "#c00a56", "#c00a56", "#c00a56", "#FF0000", "#FF0000", "#FF0000", "#FF0000",
           "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50", "#FF7F50",
           "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#F0E68C", "#32CD32", "#32CD32"]
```

```python
p.vbar(x=activos["Estado"], bottom=0, top=activos["Casos confirmados"], width=0.8, color=colores)
```

```python
p.text(x=activos["Estado"], y=activos["Casos confirmados"], text=activos["Casos confirmados"], angle=3.1416/2, text_baseline="middle", y_offset=-5)
```

```python
p.xaxis.major_label_orientation = 3.1416/2

p.yaxis.axis_label = 'Casos confirmados'

p.grid.visible = False
```

```python
bpl.show(p)
```
